# Solving Google API problem

```python
call = AudioSegment.from_wav('audio.wav')
```

```python
chunksize = 60000
```

```python
r = sr.Recognizer()
text_file = open('output.txt', 'w')

for part, i in enumerate(range(0, len(call), chunksize)):
  chunk = call[i:i + chunksize]
  filename = 'tmp/part'+str(part+1).zfill(2)+'.wav'
  chunk.export(filename, format ="wav")

  

  with sr.AudioFile(filename) as  source:
  
    audio_text = r.listen(source)

    try:
      text = r.recognize_google(audio_text)

      text_file.write(str(part+1).zfill(2) + ': ')
      text_file.write(text)
      text_file.write('\n')
      
      print('Convert')
      print(filename)
      
    except:
      print('Sorry')


text_file.close()
```
